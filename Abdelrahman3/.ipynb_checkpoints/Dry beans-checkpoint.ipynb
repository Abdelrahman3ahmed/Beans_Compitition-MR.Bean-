```python
# Data Pre-processing
```

```python
def Change_class(data):
    data1 = data.rename(columns = {"y" : "Class"} )
    data1['Class'].replace({'DERMASON':0,'SIRA':1,'SEKER':2,'HOROZ':3,'CALI':4,'BARBUNYA':5,'BOMBAY':6} ,inplace  =True)   
    return data1 
```

```python
Pre_processed_training_data = Change_class(Pure_training_data)
```

```python
Pre_processed_training_data.head()
```

```python
def Imbalanced_Visualization(data):
    # visualize the target variable
    g = sns.countplot(data['Class'])
    g.set_xticklabels(list(data['Class'].unique()))
    plt.show()
```

```python
Imbalanced_Visualization(Pre_processed_training_data)
```
